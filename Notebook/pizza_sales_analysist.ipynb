### KPI's

#### Daily Trend - Total Orders

```python

weekday_order = ['Monday','Tuesday','Wednesday','Thursday',
                 'Friday','Saturday','Sunday']

df['day_name'] = pd.to_datetime(df['order_date'], dayfirst=True).dt.day_name()

order_by_day = (
    df.groupby('day_name', observed=False)['order_id']
      .nunique()
      .reindex(weekday_order, fill_value=0)
)

ax = order_by_day.plot(kind='bar', figsize=(8,5), color='green', edgecolor='black')
plt.title('Total Orders by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Orders')
plt.xticks(rotation=45)

for i, val in enumerate(order_by_day):
    plt.text(i, val + 0.5, str(val), ha='center', va='bottom', fontweight='bold')

# plt.tight_layout()
plt.show()

```

#### Daily Trend - Total Revenue

```python
weekday_order = ['Monday','Tuesday','Wednesday','Thursday',
                 'Friday','Saturday','Sunday']

df['day_name'] = pd.to_datetime(df['order_date'], dayfirst=True).dt.day_name()

order_by_day = (
    df.groupby('day_name', observed=False)['total_price']
    .sum()
    .reindex(weekday_order, fill_value=0)
)

ax = order_by_day.plot(kind='bar', figsize=(8,5), color='purple', edgecolor='black')
plt.title('Total Revenue by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)

for i, val in enumerate(order_by_day):
    ax.text(i, val + 0.5, str(val), ha='center', va='bottom', fontsize=10, fontweight='bold')
    
plt.tight_layout()
plt.show()
```

#### Hourly Trend - Total Orders

```python
df['order_hour'] = pd.to_datetime(df['order_time'], format='%H:%M:%S').dt.hour

order_by_hour = (
    df.groupby('order_hour', observed=False)['order_id']
    .nunique()
)


ax = order_by_hour.plot(kind='bar', figsize=(10,5), color='orange', edgecolor='black')

plt.title('Total Orders by Hour of the Day')
plt.xlabel('Hour of Day (24-Hour Format)')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)

for i, val in enumerate(order_by_hour):
    ax.text(i, val + 0.5, str(val), ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

#### Hourly Trend - Total Revenue

```python
df['order_hour'] = pd.to_datetime(df['order_time'], format='%H:%M:%S').dt.hour

order_by_hour = (
    df.groupby('order_hour', observed=False)['total_price']
    .sum()
)

ax = order_by_hour.plot(kind='bar', figsize=(10,5), color='blue', edgecolor='black')

plt.title('Total Revenue by Hour of the Day')
plt.xlabel('Hour of Day (24-Hour Format)')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)

for i, val in enumerate(order_by_hour):
    ax.text(i, val + 0.5, str(val), ha='center', va='bottom', fontsize= 10, fontweight='bold')

plt.tight_layout()
plt.show()
```

#### Month Trend - Total Oder

```python
# Make sure months are ordered
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

df['month_name'] = pd.to_datetime(df['order_date'], dayfirst=True).dt.month_name()

df['month_name'] = pd.Categorical(df['month_name'], categories=month_order, ordered=True)

# Group by month and sum revenue (or count orders)
order_by_month = df.groupby('month_name', observed=False)['order_id'].nunique()

# Plot area manually
fig, ax = plt.subplots(figsize=(10,5))

# Plot the filled area
ax.fill_between(order_by_month.index, order_by_month.values, color='blue', alpha=0.3)

# Plot the line with markers
ax.plot(order_by_month.index, order_by_month.values, color='blue', linewidth=2, marker='o')

# Add value labels
for i, val in enumerate(order_by_month):
    ax.text(i, val+ 5, f"{val:,.0f}", ha='center', va='bottom', fontsize=10, fontweight='bold')

# Titles and labels
plt.title('Total Order by Month')
plt.xlabel('Month')
plt.ylabel('Total Order')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Month Trend - Total Revenue

```python
# Make sure months are ordered
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

df['month_name'] = pd.to_datetime(df['order_date'], dayfirst=True).dt.month_name()

df['month_name'] = pd.Categorical(df['month_name'], categories=month_order, ordered=True)

# Group by month and sum revenue (or count orders)
order_by_month = df.groupby('month_name', observed=False)['total_price'].sum()


# Plot area manually
fig, ax = plt.subplots(figsize=(10,5))

# Plot the filled area
ax.fill_between(order_by_month.index, order_by_month.values, color='green', alpha=0.3)

# Plot the line with markers
ax.plot(order_by_month.index, order_by_month.values, color='green', linewidth=2, marker='o')


# Add value labels
for i, val in enumerate(order_by_month):
    ax.text(i, val + 5, f"${val:,.0f}", ha='center', va='bottom', fontsize=10, fontweight='bold')

# Titles and labels
plt.title('Total Revenue by Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### % of Sales by Category


```python
category_sales= (
    df.groupby('pizza_category')['total_price']
    .sum()
    .sort_values(ascending=False)

)
# Convert to percentage 
category_pct = category_sales / category_sales.sum() * 100

# Create figure
plt.figure(figsize=(5,5))

# Color palette
colors = plt.get_cmap('tab20').colors # nic color palette

# Pie (Donut) chart
plt.pie(
    category_pct, 
    labels=category_pct.index, 
    autopct=lambda p: f'{p:.1f}%\n(${p*category_sales.sum()/100:,.0f})',
    startangle=90, 
    colors=colors, 
    wedgeprops={'edgecolor': 'black', 'width':0.4},
    explode = [0.05 if i == 0 else 0 for i in range(len(category_pct))]
        
    )


plt.title('Percentage of Sales by Pizza Category')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

#### % Sales by Pizza Size & Category

```python
# Create pivot table
sale_pivot = (
    df.pivot_table(
        index='pizza_category',
        columns='pizza_size',
        values='total_price',
        aggfunc='sum',
        fill_value=0
    )
)

# Sort categories by total sales
sale_pivot = sale_pivot.loc[sale_pivot.sum(axis=1).sort_values(ascending=False).index]

# Convert to percentage of total sales
total_sales = sale_pivot.values.sum()
sale_pct = (sale_pivot / total_sales) * 100

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    sale_pct,
    annot=True,
    fmt=".1f",
    cmap="YlOrRd",
    linewidths=0.5,
    cbar_kws={'label': 'Percentage of Total Sales (%)'}
)

plt.title("Percentage of Sales by Pizza Category and Size")
plt.xlabel("Pizza Size")
plt.ylabel("Pizza Category")
plt.show()
```

#### Total Pizza Sold by Pizza Category

```python
pizzas_by_category = (
    df.groupby('pizza_category', observed=False)['quantity']
    .sum()
    .sort_values(ascending=False)

)

colors = list(plt.get_cmap('tab20').colors)
colors = colors[:len(pizzas_by_category)]

ax = pizzas_by_category.plot(kind='bar', figsize=(8,5), color=colors, edgecolor='black')

plt.title('Total Pizzas Sold by Category')
plt.xlabel('Pizza Category')
plt.ylabel('Total Pizzas Sold')
plt.xticks(rotation=45)

for i, val in enumerate(pizzas_by_category):
    ax.text(i, val + 5, str(val), ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

#### Top 5 Best-Selling Pizza - Total Qty


```python
best_selling_pizzas = (
    df.groupby('pizza_name', observed=False)['quantity']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

ax = best_selling_pizzas.plot(
    kind='barh',          #  horizontal bar
    figsize=(8, 5),
    color='red',
    edgecolor='black'
)

plt.title('Top 5 Best-Selling Pizzas')
plt.xlabel('Total Pizzas Sold')
plt.ylabel('Pizza Name')

# Put the best seller at the top
ax.invert_yaxis()

# Add value labels at the end of bars
for i, val in enumerate(best_selling_pizzas.values):
    ax.text(val + 2, i, val, va='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
